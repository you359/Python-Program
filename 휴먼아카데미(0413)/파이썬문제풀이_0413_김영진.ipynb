# 심화문제

1) 다음 경로의 모든 dll 파일들에 대해 확장자를 제외한 순수 파일 이름을 추출
   > "C:\Windows\System32\*.dll"
2) file_data 딕셔너리에 key는 파일 이름, value는 path 값을 갖는 아이템 추가
   > 딕셔너리 구조) file_data[key] = value  
   > 예) file_data[file_name] = path # 여기서 file_name은 1)을 통해 추출된 순수 파일 이름임

```python
for idx, path in enumerate(glob("C:\Windows\System32\*.dll")):
    print("[{}]: {}".format(idx, path))
```

```python
from glob import glob # 파일 경로를 가져오는 라이브러리
file_data = {} # 딕셔너리 선언

for path in glob("C:\Windows\System32\*.dll"):
    # file_name = [여기에 알맞은 코드를 작성하여 순수 파일 이름 추출]
    file_name_with_ext = path.split("\\")[-1]
    filename = file_name_with_ext.split(".")[0]
    file_data[filename] = path

for key in file_data.keys():
    print("[{}]: {}".format(key, file_data[key]))
```
